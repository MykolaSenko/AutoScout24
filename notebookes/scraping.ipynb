Importing dependencies

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

Scraping number of cars which are available on autoscout24.

```python


url = "https://www.autoscout24.be/nl/lst?atype=C&cy=B&desc=0&sort=standard&source=homepage_search-mask&ustate=N%2CU"

response = requests.get(url)

soup = BeautifulSoup(response.content, "lxml")


print(soup.prettify())

print(response.status_code)


```

Extracting values from json data.

```python
script_tag = soup.find(id="__NEXT_DATA__")
if script_tag:
    # Extract the text content from the script tag
    script_content = script_tag.string

    # Parse the JSON data
    data = json.loads(script_content)

    # Access the numberOfResults property
    try:
        number_of_results = data['props']['pageProps']['numberOfResults']
        print(f"There are {number_of_results} cars for sale in Belgium.")  
    except KeyError as e:
        print(f"Error: {e}")
else:
    print("Script tag with ID '__NEXT_DATA__' not found.")
```

Opening the excell file with the brand data.

```python
brands_file_path = '../data/cars_brands.xlsx'
df_brands = pd.read_excel(brands_file_path)
brands_list = df_brands.values.tolist()
brands = [item for sublist in brands_list for item in sublist]
brands = [brand.lower() for brand in brands]
brands = [brand.replace("\xa0", "-") for brand in brands]
print(brands)
print(len(brands))
```

Opening csv file with zipcodes.

```python
zipcode_file_path = '../data/zipcode-belgium.csv'
df_zipcodes = pd.read_csv(zipcode_file_path)

display(df_zipcodes)
```

Creating lists from zipcodes  dataframe.

```python
list_zipcodes = df_zipcodes['postcode'].tolist()
print(list_zipcodes)
print(len(list_zipcodes))

list_cities = df_zipcodes['locality'].tolist()
print(list_cities)
print(len(list_cities))

list_lat = df_zipcodes['long'].tolist()
print(list_lat)
print(len(list_lat))

list_long = df_zipcodes['lat'].tolist()
print(list_long)
print(len(list_long))
```

Generating link for scraping


```python
result = []
for brand in brands:
    for zipcode, city, lat, lon in zip(list_zipcodes, list_cities, list_lat, list_long):
        try:
            url_page_1 = f"https://www.autoscout24.be/nl/lst/{brand}/{zipcode}-{city}?atype=C&cy=B&damaged_listing=exclude&desc=0&lat={lat}&lon={lon}&page=1&powertype=kw&search_id=1mqvtm8cw4&sort=standard&source=homepage_search-mask&ustate=N%2CU&zipr=10"
            response = requests.get(url_page_1)
            soup_page_1 = BeautifulSoup(response.content, "lxml")
            cars_amount = soup_page_1.find(class_='ListHeader_headerTitle__Hkaeg').find('span').find('span').text
            
            if cars_amount == "0":
                print(f"No cars of {brand} were found in {zipcode}{city}")
                continue
                
            else:
                pages_amount = soup_page_1.find(class_='pagination-item--disabled pagination-item--page-indicator').text
                pages = pages_amount[4:]
                print(f"There are {cars_amount} cars of {brand} on {pages} pages for sale in {zipcode}{city}")

                for page in range(1, int(pages) + 1):
                    print(f"Processing page {page}")
                    if page == 1:
                        script_tag_1 = soup_page_1.find(id="__NEXT_DATA__")
                        if script_tag:        
                            script_content_1 = script_tag_1.string
                            data_content_1 = json.loads(script_content_1)
                            try:
                                result_page_1 = data_content_1['props']['pageProps']['listings']
                                for elem in range (len(result_page_1)):
                                    result.append(result_page_1[elem])
                            except KeyError as e:
                                print(f"Error while scraping cars from pages: {e}")
                        else: 
                            print("Script tag with ID '__NEXT_DATA__' not found.")
                    else:
                        try:
                            url_page = f"https://www.autoscout24.be/nl/lst/{brand}/{zipcode}-{city}?atype=C&cy=B&damaged_listing=exclude&desc=0&lat={lat}&lon={lon}&page={page}&powertype=kw&search_id=1mqvtm8cw4&sort=standard&source=homepage_search-mask&ustate=N%2CU&zipr=10"
                            response = requests.get(url_page)
                            soup_page = BeautifulSoup(response.content, "lxml")
                            script_tag = soup_page.find(id="__NEXT_DATA__")
                            if script_tag:        
                                script_content = script_tag.string
                                data_content = json.loads(script_content)

                                try:
                                    result_page = data_content['props']['pageProps']['listings']
                                    for elem in range (len(result_page)):
                                        result.append(result_page[elem])
                                except KeyError as e:
                                    print(f"Error while scraping cars from pages: {e}")
                            print(result)
                            print(len(result))
                            
                            
                        except KeyError as e:
                            print(f"Error while scraping cars from pages: {e}")
                            
        
        except KeyError as e:
            print(f"Error while scraping pages: {e}")




```
